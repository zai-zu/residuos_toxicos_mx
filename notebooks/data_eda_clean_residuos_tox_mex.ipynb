```python
#Eliminar las ultimas tres filas
df_residuos_tox.drop(df_residuos_tox.tail(3).index, inplace=True)
df_residuos_tox.tail()
```

```python
#Eliminar columnas 2 y de la 9 a la 14
df_residuos_tox.drop(df_residuos_tox.columns[9:15], axis=1, inplace=True)
df_residuos_tox.drop(df_residuos_tox.columns[2], axis=1, inplace=True)
df_residuos_tox.head()
```

```python
df_residuos_tox.rename(columns={'ID Entidad Federaativa': 'id_ent_fed', 'Entidad federativa': 'ent_fed', 'Año' : 'anio',
       'Subsector' :'subsector', 'Categoría de generador': 'categoria_generador', 'Categoría del residuo': 'categoria_residuo',
       'Tipo de residuo' : 'type_residuo', 'Generación estimada': 'generacion_estimada_ton_anio'},inplace= True)
df_residuos_tox.head()
```

```python
# Eliminar "n/" y limpiar espacios extra en todos los campos
df_residuos_tox = df_residuos_tox.replace("n/", "", regex=True).applymap(lambda x: x.strip() if isinstance(x, str) else x)
df_residuos_tox.head()
```

```python
df_residuos_tox['generacion_estimada_ton_anio'] = df_residuos_tox['generacion_estimada_ton_anio'].str.replace(',', '.', regex=True)

df_residuos_tox.head(358)
```

```python
#Elimnar doble comas en columnas categoria_residuo y type_residuo
df_residuos_tox['categoria_residuo'] = df_residuos_tox['categoria_residuo'].str.replace(',,', ',', regex=True)
```

```python
import re

def remove_first_dot_if_length_more_than_5(x):
    """
    Si x es un string de longitud > 5, elimina el primer '.' que aparezca.
    De lo contrario, devuelve x sin modificar.
    """
    if isinstance(x, str) and len(x) > 5:
        # El parámetro count=1 en re.sub() asegura que sólo se reemplace la primera coincidencia.
        return re.sub(r"\.", "", x, count=1)
    else:
        return x
```

```python
df_residuos_tox['generacion_estimada_ton_anio'] = df_residuos_tox['generacion_estimada_ton_anio'].apply(remove_first_dot_if_length_more_than_5)
```

```python
#Convertir columna generacion_estimada_ton/anio a float
df_residuos_tox['generacion_estimada_ton_anio'] = df_residuos_tox['generacion_estimada_ton_anio'].astype(float)
```

```python
df_residuos_tox [['anio', 'id_ent_fed']] = df_residuos_tox [['anio', 'id_ent_fed']].astype(int)
```
