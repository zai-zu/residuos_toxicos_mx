```python
!pip install pandas

import pandas as pd

# Leer el CSV limpio
df_residuos_tox = pd.read_csv('../data/processed/clean_data_residuos_tox_mex.csv')

# Ver las primeras filas del DataFrame
df_residuos_tox.head()
```

```python
df_residuos_tox.info()
```

```python
df_residuos_tox.describe()
```

```python
#Agrupar por entidades estados mayores generadores de resuidos (columna generacion_estimada_ton_anio) con el método groupby
df_residuos_tox_grouped_mayor = df_residuos_tox.groupby('ent_fed').agg({'generacion_estimada_ton_anio': 'sum'}).reset_index()
# Ordenar de mayor a menor
df_residuos_tox_grouped_mayor = df_residuos_tox_grouped_mayor.sort_values(by='generacion_estimada_ton_anio', ascending=False)
# reindexar el índice desde 1
df_residuos_tox_grouped_mayor.index = range(1, len(df_residuos_tox_grouped_mayor) + 1)
# Cambiar el nombre de la columna para que sea más descriptivo
df_residuos_tox_grouped_mayor.rename(columns={'generacion_estimada_ton_anio': 'generacion_estimada_ton_2017_2022_total'}, inplace=True)
# Mostrar el DataFrame agrupado y ordenado
df_residuos_tox_grouped_mayor.to_csv('../data/processed/df_residuos_tox_grouped_mayor.csv', index=False)
df_residuos_tox_grouped_mayor.head(10)

 
```

```python
# Agrupar por entidades estados menores generadores de resuidos (columna generacion_estimada_ton_anio) con el método groupby
df_residuos_tox_grouped_menor = df_residuos_tox.groupby('ent_fed').agg({'generacion_estimada_ton_anio': 'sum'}).reset_index()
# Ordenar de menor a mayor
df_residuos_tox_grouped_menor = df_residuos_tox_grouped_menor.sort_values(by='generacion_estimada_ton_anio', ascending=True)
# reindexar el índice desde 1
df_residuos_tox_grouped_menor.index = range(1, len(df_residuos_tox_grouped_menor) + 1)
# Cambiar el nombre de la columna para que sea más descriptivo
df_residuos_tox_grouped_menor.rename(columns={'generacion_estimada_ton_anio': 'generacion_estimada_ton_2017_2022_total'}, inplace=True)
# Mostrar el DataFrame agrupado y ordenado

df_residuos_tox_grouped_menor.to_csv('../data/processed/df_residuos_tox_grouped_menor.csv', index=False)
df_residuos_tox_grouped_menor.head(10)


```

```python
#Agrupar por entidad y año con el método groupby las entidades con mayor generación de residuos (columna generacion_estimada_ton_anio) con el método groupby
df_residuos_tox_grouped_generacion_anio = df_residuos_tox.groupby(['ent_fed', 'anio']).agg({'generacion_estimada_ton_anio': 'sum'}).reset_index()
# Ordenar de mayor a menor
df_residuos_tox_grouped_generacion_anio = df_residuos_tox_grouped_generacion_anio.sort_values(by='generacion_estimada_ton_anio', ascending=False)
# reindexar el índice desde 1
df_residuos_tox_grouped_generacion_anio.index = range(1, len(df_residuos_tox_grouped_generacion_anio) + 1)
# Cambiar el nombre de la columna para que sea más descriptivo
df_residuos_tox_grouped_generacion_anio.rename(columns={'generacion_estimada_ton_anio': 'generacion_estimada_ton_anio_total'}, inplace=True)
# Guardar el DataFrame agrupado y ordenado en un archivo CSV

df_residuos_tox_grouped_generacion_anio.to_csv('../data/processed/df_residuos_tox_grouped_generacion_anio.csv', index=False)
df_residuos_tox_grouped_generacion_anio.head(20)
```
